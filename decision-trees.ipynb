```python
#Loading new data for random forests exercise
file_path2 = Path("linear_regression_salary/Resources/loans_data_encoded_3.csv")
df_loans2 = pd.read_csv(file_path2)
df_loans2.head(20)
```

```python
#Define the features set.
X = df_loans2.copy()
X = X.drop("bad", axis=1)
X.head(20)
```

```python
#Define the target set.
y = df_loans2["bad"].ravel()
y[:20]
```

```python
#Splitting into Train and Test sets.
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y, random_state=78)
```

```python
#Creating a StandardScaler instance.
scaler2 = StandardScaler()
#Fitting the Standard Scaler with the training data.
X_scaler2 = scaler2.fit(X_train2)
```

```python
#Scaling the data.
X_train_scaled2 = X_scaler2.transform(X_train2)
X_test_scaled2 = X_scaler2.transform(X_test2)
```

```python
#Creating random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78)
```

```python
rf_model
```

```python
#Fitting the model.
rf_model = rf_model.fit(X_train_scaled2, y_train2)
```

```python
#Making predictions using the testing data.
predictions2 = rf_model.predict(X_test_scaled2)
predictions2
```

```python
# Calculating the confusion matrix.
cm2 = confusion_matrix(y_test2, predictions2)

# Create a DataFrame from the confusion matrix.
cm2_df = pd.DataFrame(
    cm2, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm2_df
```

```python
# Calculating the accuracy score.
acc_score2 = accuracy_score(y_test2, predictions2)
acc_score2
```

```python
# Displaying results
print("Confusion Matrix")
display(cm2_df)
print(f"Accuracy Score : {acc_score2}")
print("Classification Report")
print(classification_report(y_test2, predictions2))
```

```python
#In summary, bootstrapping is simply a sampling technique with which a number of samples are made, 
#and in which an observation can occur multiple times.
```

```python
#In the aggregation step, different classifiers are run, using the samples drawn in the 
#bootstrapping stage. Each classifier is run independently of the others, and 
#all the results are aggregated via a voting process. Each classifier will vote 
#for a label (a prediction). The final prediction is the one with the most votes.
```

```python
#Boosting is another technique to combine weak learners into a strong learner.
#In boosting, however, the weak learners are not combined at the same time. 
#Instead, they are used sequentially, as one model learns from the mistakes of the previous model.
```
